```python
from pathlib import Path
from loguru import logger
import polars as pl

def read_input_csv(config: dict) -> pl.DataFrame:
    csv_path = config.get("csv")
    read_opts = config.get("read_csv_opts", {})

    logger.info("Reading input CSV from {}", csv_path)
    try:
        df = pl.read_csv(source=csv_path, **read_opts)
        logger.success("Successfully read CSV with {} rows and {} columns", df.height, df.width)
        return df
    except FileNotFoundError:
        logger.error("CSV file not found: {}", csv_path)
        raise
    except Exception as e:
        logger.exception("Failed to read CSV from {}: {}", csv_path, e)
        raise
```
